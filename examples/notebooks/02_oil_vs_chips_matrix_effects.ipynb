# Oil vs Chips Matrix Effects
Compare oils vs chips synthetic datasets using a multi-input protocol.

```python
import pandas as pd
from pathlib import Path
from foodspec.protocol_engine import ProtocolConfig, ProtocolRunner

oil_df = pd.read_csv('examples/data/oil_synthetic.csv')
chips_df = pd.read_csv('examples/data/chips_synthetic.csv')
proto = {
    'name': 'oil_vs_chips_demo',
    'steps': [
        {"type":"rq_analysis","params":{"oil_col":"oil_type","matrix_col":"matrix","heating_col":"heating_stage"}},
        {"type":"output","params":{"output_dir":"runs_notebook"}}
    ],
    'expected_columns': {'oil_col':'oil_type','matrix_col':'matrix','heating_col':'heating_stage'}
}
cfg = ProtocolConfig.from_dict(proto)
runner = ProtocolRunner(cfg)
res = runner.run([oil_df])
print('Run dir:', res.run_dir)
print('Tables:', list(res.tables.keys()))
print('Top oil vs chips divergence (if present):')
if 'oil_vs_chips_summary' in res.tables:
    print(res.tables['oil_vs_chips_summary'].head())
```
